```python
from google.colab import drive
import zipfile
import os

# 1. חבר את Google Drive לקולאב
drive.mount('/content/drive')

# 2. הגדר את הנתיבים
zip_file_path = '/content/drive/MyDrive/Head datasets.v2-headdatasets-v2.yolov8-obb/yael.zip'  # עדכן את הנתיב לקובץ ה-ZIP שלך
extract_dir = '/content/drive/MyDrive'    # עדכן את נתיב תקיית היעד

# 3. חלץ את הקובץ
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f'Files extracted to {extract_dir}')
```

הורדת תקיית התוצאות והמודל למחשבבבב זה הכי חשובבבב שימי לב לשים את הנתיב של תקיית

```python
from google.colab import drive
import shutil
from google.colab import files

# מיפוי גוגל דרייב
drive.mount('/content/drive')

# נתיב התקייה שברצונך להוריד ונתיב קובץ ה-zip
folder_path = '/content/runs'    # רק פה לשים ניתוב לתקיית runs שנוצרה במהלך האימון
zip_path = '/content/runs.zip'

# יצירת קובץ zip מהתקייה
shutil.make_archive(zip_path.replace('.zip', ''), 'zip', folder_path)

# הורדת קובץ ה-zip למחשב
files.download(zip_path)
```
